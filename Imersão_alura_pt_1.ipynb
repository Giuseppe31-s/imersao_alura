###6) Plotando gráficos melhores


```python
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
#utilizando seaborn
sns.set_theme(context= 'notebook')
plt.figure(figsize=(12,6))
sns.histplot(data = dados, x = 'g0', hue = 'tratamento', kde = True)
plt.title('Distribuição da expressão genetica G0', fontsize = 16);
```

```python
plt.figure(figsize=(8,8))
plt.hist(data = dados,x ='g0', bins= 50);
plt.title('Distribuição da expressão genetica G0',fontsize = 16);
```
