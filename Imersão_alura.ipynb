```python
# Definindo função heartmap
# Habilitando anotação em formato booleano
def heartmap(corr, annot = False):

# configurando gráfico de cores
# mudando fundo do gráfico
  sns.set_theme(style='white')
# gerando mask em formato de triângulo
  mask = np.triu(np.ones_like(corr,dtype = bool))
#configurando figura
  f,ax = plt.subplots(figsize = (11,9))
# Criando colormap 
  cmap = sns.diverging_palette(230,20,as_cmap=True)
# plotando colormap com configurações mask e cmap
  sns.heatmap(corr,mask=mask,cmap= cmap,center = 0,
            square = True, linewidths=.5, cbar_kws={'shrink': .5},annot = annot);
```

Mas, como interpretar esses valores? Bom, temos a seguinte divisão:

- Valores muito próximos de 1 ou -1: variáveis altamente correlacionadas
- Valores muito próximos de 0: variáveis pouco ou não correlacionadas

E, o que diferencia se essa correlação será proporcional ou inversamente proporcional, será o sinal. Quer dizer:

- Valores muito próximos de 1: variáveis proporcionalmente correlacionadas
- Valores muito próximos de -1: variáveis correlacionadas inversamente proporcionais



```python
# configurando gráfico de cores
# mudando fundo do gráfico
sns.set_theme(style='white')
# gerando mask em formato de triângulo
mask = np.triu(np.ones_like(corr,dtype = bool))
#configurando figura
f,ax = plt.subplots(figsize = (11,9))
# Criando colormap 
cmap = sns.diverging_palette(230,20,as_cmap=True)
# plotando colormap com configurações mask e cmap
sns.heatmap(corr,mask=mask,cmap= cmap,center = 0,
            square = True, linewidths=.5, cbar_kws={'shrink': .5});
```

```python
#importando as bibliotecas para treinamento do modelo
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model_R =  LogisticRegression(max_iter=1000)
```

```python
dados_resultados.head()
```

```python
#separando nossas variáveis explicativas da variável alvo
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
```

```python
#modelo de Regressão Logística criado
model_R.fit(x,y)
```

```python
#o statefy leva em consideração a proporção de true e false para uma previsão mais detalhada
# a gente precisa separar os dados para testar se o nosso modelo funciona ou não é qual séria a % de aproveitamento
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state= 42,stratify = y)
```

```python
#ajustando o modelo aos dados de treino
model_R.fit(x_train,y_train)
```

```python
#verificando a qualidade do modelo
model_R.score(x_test,y_test)
```

```python
#DummyClassifier é um classificador que faz previsões usando regras simples.
from sklearn.dummy import DummyClassifier
modelo_dummy = DummyClassifier(strategy='most_frequent')
modelo_dummy.fit(x_train, y_train)
modelo_dummy.score(x_test,y_test)
previsao_dummy = modelo_dummy.predict(x_test)
```

```python
#importando modelo de acurácia de previsão
from sklearn.metrics import accuracy_score
```

```python
#vendo a acurácia do modelo
accuracy_score(y_test, previsao_dummy)
```

```python
# vendo qual a porcentagem de verdeiros e falsos
y_test.value_counts(normalize=True)
```

```python
#importando modelo de previsão de Árvores de decisão
#Estipulando limite de produnfidade em 3
from sklearn.tree import DecisionTreeClassifier
modelo_dtc =  DecisionTreeClassifier(max_depth=3)
```

```python
#Treinando nosso modelo
modelo_dtc.fit(x_train,y_train)
previsao_dtc = modelo_dtc.predict(x_test)
accuracy_score(y_test,previsao_dtc)
```

```python
#Plotando gráfico que demostra o que nosso modelo faz
from sklearn import tree

fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_dtc, 
               ax = ax,
               fontsize=10,
               rounded=True,
               filled = True,
               feature_names = x_train.columns,
               class_names = ['Não Ativado', 'Ativado']);
```

```python
# limitando o máximo de profundidade e analisando o melhor
teste = []
treino = []

for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth=i)
  modelo_arvore.fit(x_train, y_train)
  teste.append(modelo_arvore.score(x_test,y_test))
  treino.append(modelo_arvore.score(x_train,y_train))

```

Não dá para obter uma melhor performance apenas aumentando a profundidade, a melhor performance está no 4, depois disso a maquina se acostuma com os dados, esse processo é chamado de [Overfitting](https://pt.wikipedia.org/wiki/Sobreajuste) 

```python
# plotando gráfico
sns.lineplot(x=range(1,15), y=teste, label='teste')
sns.lineplot(x=range(1,15), y=treino, label='treino');
```
